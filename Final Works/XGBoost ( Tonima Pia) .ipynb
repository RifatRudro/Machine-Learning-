```python
import pandas as pd #Works with Data Frames
import numpy as np #Numerical Py 
import matplotlib.pyplot as plt # Graph Plotting
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1)) #Scaling down the values using standard Scalar
df['scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1,1)) #Scaling down the values using standard Scalar

df.drop(['Time','Amount'], axis=1, inplace=True) #Axis = 1, Row | Axis = 0, Column

scaled_amount = df['scaled_amount']
scaled_time = df['scaled_time']

df.drop(['scaled_amount', 'scaled_time'], axis=1, inplace=True)
df.insert(0, 'scaled_amount', scaled_amount) #Inserting Columns in specific positions
df.insert(1, 'scaled_time', scaled_time) #Inserting Columns in specific positions

# Amount and Time are Scaled!

df.head() # Will show only first 5 rows
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('Class', axis=1) # Training values
y = df['Class'] # Corresponding labels

# 0 - Non fraud
# 1 - Fraud

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) 

# Total Rows = 284806
# Will be using 33% for Testing
# 284806*0.33 = 93,985.98 Test
# Train = 284806 - 93,985.98 = 190,820
```

```python
df = df.sample(frac=1)

# amount of fraud classes 492 rows.
fraud_df = df.loc[df['Class'] == 1]
non_fraud_df = df.loc[df['Class'] == 0][:492]

normal_distributed_df = pd.concat([fraud_df, non_fraud_df])

# Shuffle dataframe rows
new_df = normal_distributed_df.sample(frac=1, random_state=42)

new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
import seaborn as sns

print('Distribution of the Classes in the subsample dataset')
print(new_df['Class'].value_counts()/len(new_df))

sns.countplot('Class', data=new_df)
plt.title('Equally Distributed Classes', fontsize=14)
plt.show()
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
